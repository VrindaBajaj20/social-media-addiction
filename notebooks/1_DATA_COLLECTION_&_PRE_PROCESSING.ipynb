# **1. PROBLEM DEFINITION**


**1.1. Research Objective**

Goal: Determine whether Fear of Missing Out (FoMO) or personality traits are stronger predictors of digital addiction.

Research Question: “Which is a stronger predictor of digital addiction: Fear of Missing Out (FoMO) or personality traits?”


**1.2. Key Variables**

Predictors:

FoMO Score (Fomo)

Personality Traits (Nevrotik, Disdon, Acikfikir)

Target Variable: Social Media Addiction Score (Sosvirtol)


**1.3. Hypotheses**

Null Hypothesis (H0): Personality traits are a stronger predictor of digital addiction than FoMO.

Alternative Hypothesis (H1): FoMO is a stronger predictor of digital addiction than personality traits.

**1.4. Scope and Constraints**

Scope: Focus on the relationship between FoMO, personality traits, and digital addiction.

Constraints: Limited to the variables available in the Digital Addiction Research Dataset.

**1.5. Deliverables**

A predictive model to quantify the impact of FoMO and personality traits on digital addiction.

Statistical analysis to test hypotheses.

Visualizations and insights for stakeholders.

# **2. DATA COLLECTION**

**A. DATA LOADING**

```r
library(tidyverse)
data <- read.csv("/content/social_media.csv")
```

**B. DATA STRUCTURE**


```r
str(data)
head(data, 10)
colnames(data)
```

**C. MISSING VALUES & DUPLICATES CHECK**

```r
# missing value check
colSums(is.na(data))
colMeans(is.na(data)) * 100

# duplicates check
sum(duplicated(data))
data <- data %>% distinct()
```

```r
summary(data)
```

# **3. DATA PREPARATION & PROCESSING**

**FEATURE SELECTION**


```r
# Scope of discussion is limited to FOMO vs Personality Traits
# applied renaming to english for selected features
data <- data %>% select(Fomo, Nevrotik, Disdon, Acikfikir, Sosvirtol) %>%
              rename(FoMO = Fomo, # Renaming the selected column to 'FoMO'
                     Neuroticism = Nevrotik,
                     Disinhibition = Disdon,
                     Openness = Acikfikir,
                     SocialMediaAddiction = Sosvirtol)
str(data)
write.csv(data, file = "social_media_1.csv", row.names = FALSE)
```

**DATA DISTRIBUTION**

```r
library(ggplot2)
ggplot(data, aes(x = FoMO)) + geom_histogram(binwidth = 0.5, fill = "blue") + ggtitle("Distribution of FoMO")
ggplot(data, aes(x = Neuroticism)) + geom_histogram(binwidth = 0.5, fill = "green") + ggtitle("Distribution of Neuroticism")
ggplot(data, aes(x = Disinhibition)) + geom_histogram(binwidth = 0.5, fill = "red") + ggtitle("Distribution of Disinhibition")
ggplot(data, aes(x = Openness)) + geom_histogram(binwidth = 0.5, fill = "purple") + ggtitle("Distribution of Openness")
```

image.png

**DATA STANDARDIZATION(Z-score)**

For Neuroticism, Disinhibition, and FoMO(ML)

```r
library(dplyr)
data<- data %>%
  mutate(across(c(FoMO, Neuroticism, Disinhibition), scale))
```

**DATA NORMALIZATION**

For Openness and FoMO(DL)

```r
data<- data %>%
  mutate(across(c(FoMO, Openness), ~ (. - min(.)) / (max(.) - min(.))))
```

```r
write.csv(data, file = "social_media_cleaned.csv", row.names = FALSE)
```
